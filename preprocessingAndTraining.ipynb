```python
# replacing null value values with dominating trait nucleiotide.
def replacingnull(value,alleles):
  if pd.isna(value):
    letterbefore=alleles.split("/")[0]
    return letterbefore*2
  else:
    return value
```

```python
for column in genoDf.columns[1:]:
    for index, row in genoDf.iterrows():
        genoDf.at[index, column] = replacingnull(row[column], row['alleles'])

genoDf = genoDf.drop(['alleles'], axis=1)
```

```python
# Finding the correlation using the snp ratio and/or pooled and finding the number of common snps which highly influence the majority snp
p=[]
f=[]
pooled=['DH_Pooled','GFD_Pooled','GNPS_Pooled','GWPS_Pooled','PH_Pooled','GY_Pooled']
md=pd.read_csv("/content/datasets/merged_dataset.csv")

snp_data = md.iloc[:, 1:3530]

for i in pooled:
    gfd_pooled = md[i]
    corr_gfd = snp_data.corrwith(gfd_pooled)
    snp_corr_df = pd.DataFrame({'SNP': snp_data.columns, 'Correlation':corr_gfd})
    snp_corr_df = snp_corr_df.sort_values(by='Correlation', ascending=False)

    k=[]

    for i,j in enumerate(snp_corr_df['Correlation']):
        if(j>0):
            k.append(snp_corr_df['SNP'].iloc[i])

    # k.sort(reverse=True)
    k = k[0:len(k)*4//5]
    p.append(k)

for i in p:
    print(len(i))
f = p
list1 = f[0]
list2 = f[1]
list3 = f[2]
list4 = f[3]
list5 = f[4]
list6 = f[5]

set1=set(list1)
set2=set(list2)
set3=set(list3)
set4=set(list4)
set5=set(list5)
set6=set(list6)

common_elements=set1.intersection(set2,set3,set4,set5,set6)
common_elements_list=list(common_elements)
print(len(common_elements_list))
'''You may get a different number of SNPs;
I got 24 SNPs for the first time that influence the traits the most.
After running it again, I got 5 SNPs.
It may differ sometimes, so don't worry; that will be fine.
Change the code according to the number of SNPs you got.
I trained using a dataset with 24 SNPs.'''
```

```python
# Removing the non related snps from the dataset
dp=[]
for i in sorted(md.columns[1:]):
    if i not in common_elements_list:
        dp.append(i)

pooled=['DH_Pooled', 'GFD_Pooled', 'GNPS_Pooled', 'GWPS_Pooled', 'PH_Pooled', 'GY_Pooled']
for i in pooled:
    dp.remove(i)
```

```python
#without scaling the data
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import pickle

# Load your dataset
df = md1

# x should be the genetic sequences
# y should be the trait values assuming the DNA sequences start from the second column
x = df.iloc[:, 1:25]

# Extract the trait values
y = df[['DH_Pooled', 'GFD_Pooled', 'GNPS_Pooled', 'GWPS_Pooled', 'PH_Pooled', 'GY_Pooled']]

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=123)
# Print the highest and lowest values for each trait in the training set
print("Training Set:")
for trait in y.columns:
    max_value_train = y_train[trait].max()
    min_value_train = y_train[trait].min()
    print(f'Trait: {trait}')
    print(f'Highest Value in Training Set: {max_value_train}')
    print(f'Lowest Value in Training Set: {min_value_train}')
    print()

# Print the highest and lowest values for each trait in the test set
print("Test Set:")
for trait in y.columns:
    max_value_test = y_test[trait].max()
    min_value_test = y_test[trait].min()
    print(f'Trait: {trait}')
    print(f'Highest Value in Test Set: {max_value_test}')
    print(f'Lowest Value in Test Set: {min_value_test}')
    print()

# Build and train a random forest regression model
random_forest_model2 = RandomForestRegressor(n_estimators=500, random_state=123)
random_forest_model2.fit(X_train, y_train)
with open('random_forest_model2.pkl', 'wb') as model_file:
    pickle.dump(random_forest_model2, model_file)
y_pred = random_forest_model2.predict(X_test)

# Create an empty list to store RMSE values
rmse_values = []

# Iterate over each trait
traits = ['DH_Pooled', 'GFD_Pooled', 'GNPS_Pooled', 'GWPS_Pooled', 'PH_Pooled', 'GY_Pooled']

for trait in traits:
    trait_index = y.columns.get_loc(trait)

    # Extract actual and predicted values for the current trait
    y_actual_trait = y_test[trait]
    y_pred_trait = y_pred[:, trait_index]

    # Calculate Mean Squared Error for the current trait
    mse_trait = mean_squared_error(y_actual_trait, y_pred_trait)
    rmse_trait = np.sqrt(mse_trait)

    rmse_values.append(rmse_trait)

    print(f'RMSE for {trait}: {rmse_trait}')

# Now you can use the rmse_values list as needed in your analysis.
print(rmse_values)
print(sum(rmse_values) / len(rmse_values))
```

```python
def map_sequence_to_numbers(sequence, geno_map):
    # Check if the sequence length is even
    if len(sequence) % 2 != 0:
        raise ValueError("The sequence length must be even.")

    # Split the sequence into chunks of two characters
    chunks = [sequence[i:i+2] for i in range(0, len(sequence), 2)]

    # Map each chunk to its corresponding number using the geno_map
    mapped_numbers = [geno_map[chunk] for chunk in chunks if chunk in geno_map]

    return mapped_numbers

# Example usage
geno_map = {'AA':1, 'AT':2, 'AG':3, 'AC':4, 'TT':5, 'TG':6, 'TC':7, 'GG':8, 'CG':9, 'CC':10}
sequence = "GGCCGGCCGGTTTGGGGGTTCCGGGGTTGGCCGGGGAGCCCCCCCCCC"  # Example input
mapped_numbers = map_sequence_to_numbers(sequence, geno_map)



# Generate and predict a new sequence
new_sequence_1 = np.array(mapped_numbers).reshape(1, -1)  # Reshape to match input shape
# scaled_new_sequence = scaler_x.transform(new_sequence_1)
predictions = random_forest_model2.predict(new_sequence_1)

# Transform transform the scale of predictions to get the original scale
# predictions = scaler_y.inverse_transform(predictions_scaled)

# print("Root Mean Squared Error (MSE):", rmse)
print(rmse_values)
print("Predictions:", predictions)
```


# **FEED FORWARD NEURAL NETWORK (FFNN)**


```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error

# Define the genotype to number mapping
geno_map = {'AA': 1, 'AT': 2, 'AG': 3, 'AC': 4, 'TT': 5, 'TG': 6, 'TC': 7, 'GG': 8, 'CG': 9, 'CC': 10}

# Function to convert genotype sequence to numerical format
def convert_genotypes(sequence):
    genotype_list = [sequence[i:i+2] for i in range(0, len(sequence), 2)]
    number_list = [geno_map.get(genotype, 0) for genotype in genotype_list]
    return np.array(number_list).reshape(1, -1)

def create_model(input_shape, num_outputs):
    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(64, activation='relu', input_shape=input_shape),
        tf.keras.layers.Dense(32, activation='relu'),
        tf.keras.layers.Dense(num_outputs)  # Adjust the number of output neurons
    ])
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

num_outputs = y_train.shape[1]
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=123)
scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()
X_train_scaled = scaler_x.fit_transform(X_train)
X_test_scaled = scaler_x.transform(X_test)
y_train_scaled = scaler_y.fit_transform(y_train)
y_test_scaled = scaler_y.transform(y_test)

learning_rate = 0.001
batch_size = 32
epochs = 100

print("X_train_scaled shape:", X_train_scaled.shape)
print("y_train_scaled shape:", y_train_scaled.shape)

# Define early stopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True)

# Create and train the model with early stopping
model = create_model(input_shape=(24,), num_outputs=num_outputs)
history = model.fit(X_train_scaled, y_train_scaled, epochs=300, batch_size=batch_size,
                    validation_split=0.2, callbacks=[early_stopping])
print("number of epochs ran:",len(history.history['loss']))
# Predict and evaluate
user_input ="GGCCGGCCGGTTTGGGGGTTCCGGGGTTGGCCGGGGAGCCCCCCCCCC"
new_sequence = convert_genotypes(user_input)
scaled_new_sequence = scaler_x.transform(new_sequence)
predictions_scaled = model.predict(scaled_new_sequence)
print("Predictions_scaled:", predictions_scaled)
predictions = scaler_y.inverse_transform(predictions_scaled)
print("Predictions:", predictions)

# Calculate Root Mean Squared Error on test data
predictions_test = model.predict(X_test_scaled)
rmse = np.sqrt(mean_squared_error(y_test_scaled, predictions_test))
print("Root Mean Squared Error (MSE) on Test Data:", rmse)
```
